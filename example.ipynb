```python
from madoka import Figure

fig = Figure((10, 10), (3, 2))
```

```python
# x and y
fig.plot([0, 1, 3], [1, 2, 6])
fig.add_grid().show()
```

```python
# increment to the next subplot
fig.next()

# y only (x will be inferred)
# Multiple y values 
fig.plot([[1.0, 2.0, 3.0], 
          [1.1, 2.4, 3.6],
          [0.9, 2.2, 4.0]],
         fill_alpha=0.5)
fig.add_grid().show()
```

```python
fig.next()
fig.bar([1, 2, 3], [4, 5, 1],
        labels=['male', 'female'],
        add_annotate=True,
        annotate_offset=0.1,
        annotate_fontsize=10)
fig.legend(fontsize=10)
fig.add_grid(axis='y').show()
```

```python
fig.next()
fig.bar([[1, 2, 3], [1.1, 2.2, 3.5], [1, 2, 3]],
        [[4, 5, 1], [2.0, 3.0, 5.0], [5, 4, 1]],
         [[4, 5, 1], [4.2, 3.1, 4.9], [4, 5, 1]],
        tick_labels=['1', '2', '3'],
        add_annotate=True,
        annotate_offset=0.1,
        annotate_fontsize=10,
        err_alpha=0.4)
fig.add_grid(axis='y').show()
```

```python
fig.next()
fig.scatter([1, 2, 4], [3, 1, 5])
fig.scatter([3, 1, 5], [1, 2, 4])
fig.add_grid().show()
```

```python
fig.next()
fig.barh([1, 2, 3], [4, 5, 1],
        labels=['male', 'female'],
        add_annotate=True,
        annotate_offset=0.1,
        annotate_fontsize=10)
fig.legend(fontsize=10)
fig.add_grid(axis='x').show()


```
