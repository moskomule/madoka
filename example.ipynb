```python
from madoka import Figure

fig = Figure((10, 10), (2, 2))
```

```python
fig.plot([1, 2, 6])
fig.add_grid().show()
```

```python
fig.next()
fig.plot([[1.0, 2.0, 3.0], 
          [1.1, 2.4, 3.6],
          [0.9, 2.2, 4.0]])
fig.add_grid().show()
```

```python
fig.next()
fig.bar([[1, 2, 3], [1.1, 2.2, 3.5], [1, 2, 3]],
        [[4, 5, 1], [2, 3, 5], [5, 4, 1]])
fig.add_grid().show()
```

```python
fig.next()
fig.scatter([1, 2, 4], [3, 1, 5])
fig.add_grid().show()
```
